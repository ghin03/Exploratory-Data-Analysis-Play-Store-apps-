Now since all missing values in each column are handeled. Its time to standardize all the values in each column.

````python
# Let's check the values under the size column
playstore_data['Size']
````

We can see "Size" column has different units. Some are in MB and others are  in KB. so, we need to convert them into single unit.


````python
# cleaning and converting 'Size' to numeric
def clean_size(x):
    if 'Varies with device' in str(x):
        return np.nan
    elif 'k' in str(x):
        return float(str(x).replace('k', '')) / 1024
    else:
        return float(str(x).replace('M', ''))


playstore_data['Size'] = playstore_data['Size'].apply(clean_size)
````

````python
# checking the values in Price column in play store data
playstore_data['Price'].value_counts()
````

We can see that in some value $ sign is present while in other its not present. SO, We need to remove this sign to standersize the values.

````python
# remove dollar sign from Price
def remove_from_price(value):
    if type(value) == str and '$' in value:
        value = value.replace('$', '')
    return value

playstore_data['Price'] = playstore_data['Price'].apply(remove_from_price)
````

````python
#checking the values in Installs column in play_store_data
playstore_data['Installs'].value_counts()
````

In this "Installs" we have to remove these signs: +(plus) and ,(comma)

````python
#remove +(plus) and ,(comma) from Installs
def remove_from_install(values):
    if type(values) == str:
        values = values.replace(',', '').replace('+', '')
    return values

playstore_data['Installs'] = playstore_data['Installs'].apply(remove_from_install)
````

````python
#converting 'Installs' to numeric data type
playstore_data['Installs'] = playstore_data['Installs'].replace('[^\d]', '', regex=True).astype(float)
````

````python
#converting 'Reviews' to numeric data type
playstore_data['Reviews'] = playstore_data['Reviews'].astype(float)
````

````python
# converting 'Price' to numeric data type
playstore_data['Price'] = playstore_data['Price'].replace('[^\d\.]', '', regex=True).astype(float)
````

````python
#Let's check the duplicate value in "App" column in play store data
playstore_data['App'].duplicated().sum()
````

````python
# dropping duplicates value in "App" column from play store data
playstore_data.drop_duplicates(subset='App',inplace=True)
````

•All the missing values are handled.
•All columns values are standerdized by removing unwanted characters and converted in suitable datatypes.•All duplicated values have been handled.

#### Chart - 4 - Total Revenue Generated by each 'Category'

````python
# merge the data on the app
merged_data = pd.merge(playstore_data, userreviews_data, on='App')

# calculating the revenue by each category
merged_data['Revenue'] = merged_data['Installs'] * merged_data['Price']
revenue_by_category = merged_data.groupby('Category')['Revenue'].sum()
# convert revenue to million
revenue_by_category /= 1000000
# Plot the data
plt.figure(figsize=(10, 6))
revenue_by_category.plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Total Revenue (Millions)')
plt.title('Total Revenue Generated by Category')
plt.xticks(rotation=90)  # Rotate category labels for better readability
plt.tight_layout()
# adding actual revenue values to the bar chart
for i, val in enumerate(revenue_by_category.values):
    plt.annotate("{:.2f}".format(val), (i, val), ha='center', va='bottom', fontsize=8)
plt.show()
````

The 'Category' column represents categorical data, with each category being distinct and non-numeric. Bar charts are well-suited for visualizing categorical data, with each category represented by a separate bar.

#### Chart - 6 - proportion of apps that are updated frequently

````python
# converting "Last Updated" column to datetime format
playstore_data["Last Updated"] = pd.to_datetime(playstore_data["Last Updated"])
cutoff_date = pd.Timestamp.now() - pd.DateOffset(months=2)

# Filter the dataset to include only apps updated within the last 2 months
frequently_updated_apps = playstore_data[playstore_data['Last Updated'] > cutoff_date]

# Calculate the percentage of frequently updated apps
percentage_frequently_updated = (len(frequently_updated_apps) / len(playstore_data)) * 100

# Create a pie chart
labels = ['Frequently Updated', 'Not Updated in last 60 days']
sizes = [percentage_frequently_updated, 100 - percentage_frequently_updated]
colors = ['skyblue', 'lightgray']
explode = (0.1, 0)  # Explode the first slice (Frequently Updated)

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Percentage of Apps Updated Frequently')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
````

##### 1. Why did you pick the specific chart?

Pie charts are a useful tool for visualizing the distribution of a particular variable, such as the percentage of regularly updated apps, as a component of a larger image. It makes it simple for viewer to contrast the percentage of regularly updated apps with those that are not.
